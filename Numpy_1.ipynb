```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.,5.,0.2)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
#t t red -- t t**2 blue square t t**3 green ^
plt.show()
```

```python
import numpy as np
#0에서 5사이를 10번 뽑는다.
print(np.random.choice(6,10))
#0~9 를 6개르 뽑는다 replace=False이면 중복 제거
print(np.random.choice(10,6,replace=False))
#0~5까지 p=[]로 각 확률을 조정 0.1-> 10% 0.2->20% 도합 1이 되야함
print(np.random.choice(6,10, p=[0.1,0.2,0.3,0.2,0.1,0.1]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

#0~5 10번
dice = np.random.choice(6,10000,p=[0.1,0.2,0.3,0.2,0.1,0.1])

plt.hist(dice,bins=6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#10 ~ 100 200개 만듬
x = np.random.randint(10,100,500)
y = np.random.randint(10,100,500)
#0~1 100개 나옴 그 값을 *100 한다.
size = np.random.rand(500)*100

plt.scatter(x,y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
