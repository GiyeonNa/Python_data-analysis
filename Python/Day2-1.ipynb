```python
import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10,20,30,40], label='asc')
plt.plot([40,30,20,10], label='desc')
plt.legend(loc=3)#옆 서식을 표시 (loc=방향(int)) 7==5==0
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color')
plt.plot([10,20,30,40,], color='skyblue',label='skyblue')
plt.plot([40,30,20,10], color='red', label='red')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker')
plt.plot([10,20,30,40],'r ',label='circle')
plt.plot([40,30,20,10],'g^', label='triangle up')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle')
plt.plot([10,20,30,40],'r', linestyle='--', label='dashed')
plt.plot([40,2,20,10],'g', ls=':', label='dotted')
plt.plot([5,15,25,35], 'y-', label='lined')
plt.legend(loc=7)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle')
plt.plot([10,20,30,40],'r.-', label='line') #마커 . 선 line
plt.plot([40,2,20,10], 'g^--',  label='dash') # 마서 삼각형 선 dash
plt.plot([5,15,25,35], 'cs:', label='dott') #마커 네모 선 dott
plt.legend(loc=7)
plt.show()
```
