```python
import csv
import matplotlib.pyplot as plt
f = open('weather_report.csv')
data = csv.reader(f)
next(data)
result = []

#최고기온 
#중간에 없을경우
for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

plt.plot(result,'r')
plt.show()
    
```
