```python
import matplotlib.pyplot as plt
plt.scatter([1,2,3,4], [10,20,30,40], s=[100,200,400,600], c=['red','blue','green','gold'])
# s는 크기를 조정하는 코드
# c는 색상을 조정하는 코드
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter([1,2,3,4,5], [10,30,20,40,25],s=[30,60,90,120,150], c=range(5), cmap='brg')
plt.colorbar()
#colorbar() 화면에 컬러바를 추가합니다.
# cmap 컬러맵 cmap='jet', cmap='flag', cmap='brg'등 여러가지 컬러바를 가진다
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(0,100))
    y.append(random.randint(0,100))
    size.append(random.randint(10,150))

plt.scatter(x,y,s=size, c=size, cmap='jet', alpha = 0.3)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input("검색할 지역을 입력해주세요 : ")

for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break
        

plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5), dpi=300)
plt.title(name+' 지역 성별 인구 그래프')
plt.xlabel('남성 인구 수')
plt.ylabel('여성 인구 수')
plt.plot(range(max(m)), range(max(m)), 'g')
plt.scatter(m, f, c=range(101), cmap='jet')
plt.colorbar()
plt.show()
```
