```python
import csv
f = open('age.csv')
data = csv.reader(f)
line = 0

#3: 0세 이상부터 
for row in data:
    if '주안8동' in row[0]:
        print('총' + str(len(row[3:]))+ '개 구간') 
        for i in row[3:]:
            print('%3d세 : ' %line, end ='')
            line += 1
            print(i)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
line = 0

num = []

#3: 0세 이상부터 
for row in data:
    if '주안8동' in row[0]:
        print('총' + str(len(row[3:]))+ '개 구간') 
        for i in row[3:]:
            num.append(float(i))
            

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('인천광역시 미추홀구 문학동의 통계')
#figsize를 사용하면 새로운 창이 생김
plt.figure(figsize=(10,2),dpi=300) 
plt.plot(num)
plt.show()
```
