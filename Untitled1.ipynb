```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)

result = []
for row in data:
    if '주안8동' in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
plt.figure(figsize=(10,2),dpi=300)
plt.bar(range(101),result)
plt.plot(result, 'cx--')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)


result = []
result1 = []
for row in data:
    if '신림동' in row[0]:
        for i in row[3:]:
            result.append(-int(i.replace(',','')))
            
    elif '아름동' in row[0]:
        for i in row[3:]:
            result1.append(int(i.replace(',','')))


            

plt.figure(dpi=150)
plt.barh(range(101),result)
plt.barh(range(101),result1)
plt.show()
```
