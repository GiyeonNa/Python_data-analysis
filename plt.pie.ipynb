```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

area = input('검색할 동을 입력하시오 : ')
result = []


for row in data:
    if area in row[0]:
        m = 0
        f= 0 
        for i in range(101):
            m += int(row[i+3].replace(',',''))
            f += int(row[i+106].replace(',',''))
            break

result.append(m)
result.append(f)


plt.rc('font', family = 'Malgun Gothic')
color = ['crimson','darkcyan']
plt.axis('equal')

plt.pie(result, labels=['남','여'], autopct='%.1f%%', colors=color, startangle=90) #startangel 
plt.title(area+' 지역의 남녀 성비')
plt.show()
        
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
name = input('동네 입력 : ')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break
            

plt.plot(m, label = 'Male')
plt.plot(f, label = 'Female')
plt.legend()
plt.show()
    
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
result = []
name = input('궁금한 동네를 입력하주세요 : ')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
            #result.append(int(row[i].replace(',','') - int(row[i+103].replace(',','')))
        break
        
        
import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```
